### Data Ingestion

```python
import pandas as pd
import requests

url = "https://huggingface.co/datasets/San0160/CICIDS-2017/resolve/main/Wednesday-workingHours.pcap_ISCX.csv"

# stream download to avoid incomplete read
response = requests.get(url, stream=True)
with open("wednesday.csv", "wb") as f:
    for chunk in response.iter_content(chunk_size=8192):
        f.write(chunk)

# read locally
df = pd.read_csv("wednesday.csv")
df.columns = df.columns.str.strip()

print(df.shape)
print(df['Label'].value_counts())
```
